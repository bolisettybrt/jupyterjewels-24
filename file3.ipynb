```python
# 54 Using dischargeDay values print even if it even number and odd if it odd number
import pandas as pd
Hospital_Discharge=pd.read_excel('Cardiac_Outcomes.xlsx', sheet_name='Hospitalization_Discharge')

discharge_days=Hospital_Discharge["dischargeDay"]
for x in discharge_days:
  if x % 2==0:
   print(x,"Even",)
  else:
   print(x,"Odd")


    
```
